### genres
This is the type (action, adventure etc.) of game. All of them are strings. One game can have multiple types at the same time. Indie games are the most.

```python
df["genres"]
```

```python
types = []
for i in range(27075):
    for j in range(len(df["genres"][i].split(";"))):
        if df["genres"][i].split(";")[j] not in types:
            types.append(df["genres"][i].split(";")[j])
print(types)
```

```python
for j in range(len(types)):
    num = 0
    for i in range(27075):
        if types[j] in df["genres"][i]:
            num = num + 1
    print(f'number of {types[j]} games: {num}')
```
